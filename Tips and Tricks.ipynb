### Challenge of Training Generative Adversarial Networks

GANs are difficult to train.

The reason they are difficult to train is that both the generator model and the discriminator model are trained simultaneously in a game. This means that improvements to one model come at the expense of the other model.

The goal of training two models involves finding a point of equilibrium between the two competing concerns.

1. **Generator is lagging behind Discriminator**
    
    Discriminator loss is decreasing rapidly, and sometimes even going to zero (*Failure Mode!*). 


2. **Unstable Training**
    
    Since the improvement in one network comes at the expense of the other network, it is expected to see some cyclic structure in the loss values of the network. But, you might not see this structure due to the exploading or vanishing gradient in either network.


3. **Generator oscillates between generating specific examples in the domain**

    A common problem that might occur in the training of the GANs, is the generation of few specific examples over every set of the input noise. This problem is formally known as the *Mode Collapse*.

### Few Tips and Tricks 

Here are few tips and trics that I observed in my project.

1. **Use a simpler loss function**
    
    Try to use a simpler loss function in the beginning, preferrably *Mean Squared Loss* or *Binary Cross Entropy Loss*, and see how it works. 
    
    If the training and the loss values starts coming back on the track, and improve your loss functions iteratively.

2. **Use Spectral Normalization (ALWAYS!)**
    
    Spectral normalization stabilizes the training of discriminators (critics) in Generative Adversarial Networks (GANs).

    image.png

    Reference: [Spectral Normalization for GANs](https://arxiv.org/pdf/1802.05957.pdf)

 3. **Two-Timescale Update Rule (TTUR)**

    It consists of providing different learning rates for optimizing the generator and discriminator.

    For example, the discriminator trains with a learning rate 4 times greater than G. A larger learning rate means that the discriminator will absorb a larger part of the gradient signal. Hence, a higher learning rate eases the problem of slow learning of the regularized discriminator. 

    Reference: [GANs Trained by a Two Time-Scale Update Rule Converge to a Local Nash Equilibrium](https://arxiv.org/abs/1706.08500)

4. **Train with labels**

    Using labels in any way, shape or form almost always results in a dramatic improvement in the subjective quality of the samples generated by the model.

5. **Use Soft and Noisy Labels (Label Smoothing)**

    - Use Label Smoothing, i.e. if you have two target labels: Real=1 and Fake=0, then for each incoming sample, if it is real, then replace the label with a random number between 0.7 and 1.2, and if it is a fake sample, replace it with 0.0 and 0.3.

    - Make the labels the noisy for the discriminator or occasionally flip the labels when training the discriminator.

6. **Use Dropouts**

    Providing noise in form of dropouts helps to stablize the training process.


7. **Start by solving for a smaller problem**

    Start with a smaller subset of the problem, and then use transfer learning to move to the bigger problem. This little technique helps in Mode Collapse greatly, if applicable to the type of the problem.
